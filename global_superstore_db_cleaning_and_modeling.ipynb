1. Products table

```python
%%sql
CREATE TABLE Product(
    Product_ID VARCHAR(20) PRIMARY KEY,
    Product_name VARCHAR(160),
    Product VARCHAR(80),
    Product_type VARCHAR(80),
    Category VARCHAR(50),
    Sub_category VARCHAR(50)
    )
;
```

Let populate this table

```python
%%sql
INSERT INTO product (
    product_id,
    product_name,
    category,
    sub_category,
    product,
    product_type
)
SELECT
    s.Product_ID,
    MAX(s.Product_Name)  AS product_name, --collapses attribute variations
    MAX(s.Category)      AS category,
    MAX(s.Sub_Category)  AS sub_category,
    MAX(s.product)       AS product,
    MAX(s.product_type)  AS product_type
FROM Superstore s
LEFT JOIN product p --prevents re-inserts
    ON s.Product_ID = p.product_id
WHERE s.Product_ID IS NOT NULL
  AND p.product_id IS NULL
GROUP BY s.Product_ID; --1 row per product
```

6. Fact table


this table will syore measurable business events

```python
%%sql
CREATE TABLE fact_sales (
    sales_id INT AUTO_INCREMENT PRIMARY KEY,
    order_id VARCHAR(20) NOT NULL,
    customer_id VARCHAR(20) NOT NULL,
    product_id VARCHAR(20) NOT NULL,
    location_id INT NOT NULL,
    order_date_id DATE NOT NULL,
    ship_date_id DATE NOT NULL,
    ship_mode_id INT NOT NULL,
    sales DECIMAL(10,3) NOT NULL,
    quantity INT NOT NULL,
    discount DECIMAL(4,2) NOT NULL,
    profit DECIMAL(10,5) NOT NULL,
    shipping_cost DECIMAL(10,5) NOT NULL,
    order_priority VARCHAR(20),

    --constraints
    FOREIGN KEY (customer_id) REFERENCES customer(customer_id),
    FOREIGN KEY (product_id) REFERENCES product(product_id),
    FOREIGN KEY (location_id) REFERENCES location(location_id),
    FOREIGN KEY (order_date_id) REFERENCES date(date_id),
    FOREIGN KEY (ship_date_id) REFERENCES date(date_id),
    FOREIGN KEY (ship_mode_id) REFERENCES shipping(ship_mode_id)
);
```

Let populate the fact table

```python
%%sql
INSERT INTO Fact_sales(
order_id ,
customer_id ,
product_id ,
location_id ,
order_date_id ,
ship_date_id  ,
ship_mode_id ,
sales ,
quantity ,
discount ,
profit ,
shipping_cost ,
order_priority)
SELECT
    ss.Order_ID,
    ss.Customer_ID,
    ss.Product_ID,
    l.location_id,
    ss.Order_Date,
    ss.Ship_Date,
    s.ship_mode_id,
    ss.Sales,
    ss.Quantity,
    ss.Discount,
    ss.Profit,
    ss.Shipping_Cost,
    ss.Order_Priority
FROM superstore ss
JOIN customer c
    ON ss.Customer_ID = c.customer_id
JOIN product p
    ON ss.Product_ID = p.product_id
JOIN location l
    ON ss.City = l.city
   AND ss.Country = l.country
   AND (ss.State = l.state OR ss.State IS NULL)
JOIN shipping s
    ON ss.Ship_Mode = s.ship_mode
JOIN date d
    ON ss.Order_Date = d.date_id
JOIN date dd
    ON ss.Ship_Date = dd.date_id;
```

Let create a unique one table for location and delete other location tables that contained duplicates

```python
%%sql
CREATE TABLE location_cleaned AS
SELECT
    MIN(location_id) AS location_id,  -- keeps one surrogate key per group
    city,
    state,
    country,
    MIN(postal_code) AS postal_code,  -- pick one postal code per location
    MIN(market) AS market,
    MIN(region) AS region
FROM location
GROUP BY city, state, country;
```

```python
%%sql
SELECT city, state, country, COUNT(*)
FROM location_clean
GROUP BY city, state, country
HAVING COUNT(*) > 1;
```
